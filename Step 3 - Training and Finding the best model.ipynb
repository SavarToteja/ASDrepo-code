```python
label_column = "Preferred Education"

base_location = "/content/drive/Shareddrives/1:1_Savar_Toteja/Datasets/"
```

```python
import pandas as pd

train_data = pd.read_csv(base_location + "train.csv")

y_train = train_data[label_column]
X_train = train_data.loc[:, train_data.columns != label_column]
```

```python
from sklearn.model_selection import cross_val_score

# Function for K-Fold cross validation
def cross_validate(model):
    scores = cross_val_score(model, X_train, y_train, cv=5)

    print("Cross Validation accuracy Scores: ", scores)
    print("Average CV accuracy Score: ", scores.mean())
```

Training with KNN


```python
from sklearn.neighbors import KNeighborsClassifier

# Use 5-fold cross validation for hyper-parameter tuning
# Try out different values and choose the best hyper-parameters (with highest average accuracy)
knn = KNeighborsClassifier(n_neighbors=20)

cross_validate(knn)
```

Saving best KNN model

```python
import pickle

# Model with the highest accuracy
best_model_knn = KNeighborsClassifier(n_neighbors=9)
# Encode label column (in alphabetical order)
y_train = pd.factorize(y_train, sort=True)[0]

# Fit the model on the full training dataset
best_model_knn.fit(X_train, y_train)

fh = open("/content/drive/Shareddrives/1:1_Savar_Toteja/Results/" + "best_model_knn", "wb")
pickle.dump(best_model_knn, fh)
fh.close()
```

Training with Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# Use 5-fold cross validation for hyper-parameter tuning
# Try out different values and choose the best hyper-parameters (with highest average accuracy)
decision_tree = DecisionTreeClassifier(max_depth=4)

cross_validate(decision_tree)
```

Saving best Decision Tree model

```python
import pickle

# Model with the highest accuracy
best_model_dt = DecisionTreeClassifier(max_depth=7)
# Encode label column (in alphabetical order)
y_train = pd.factorize(y_train, sort=True)[0]

# Fit the model on the full training dataset
best_model_dt.fit(X_train, y_train)

fh = open("/content/drive/Shareddrives/1:1_Savar_Toteja/Results/" + "best_model_dt", "wb")
pickle.dump(best_model_dt, fh)
fh.close()
```

Training with Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Use 5-fold cross validation for hyper-parameter tuning
# Try out different values and choose the best hyper-parameters (with highest average accuracy)
random_forest = RandomForestClassifier(n_estimators=100, max_depth=7)
cross_validate(random_forest)
```

Saving best Random Forest model

```python
import pickle

# Model with the highest accuracy
best_model_rf = RandomForestClassifier(n_estimators=50, max_depth=7)
# Encode label column (in alphabetical order)
y_train = pd.factorize(y_train, sort=True)[0]

# Fit the model on the full training dataset
best_model_rf.fit(X_train, y_train)

fh = open("/content/drive/Shareddrives/1:1_Savar_Toteja/Results/" + "best_model_rf", "wb")
pickle.dump(best_model_rf, fh)
fh.close()
```

Training with MLP
